```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('characters.csv')
data
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#Split into features and target
from sklearn.model_selection import train_test_split
features = data.columns
features = features.drop('result')
target = data['result']

# Split data into test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data[features], target, test_size=0.3, random_state=12345)

# Normalize Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_nor = scaler.fit_transform(X_train)
X_test_nor = scaler.fit_transform(X_test)
```

```python
#Use LogisticRegression to classify predict
from sklearn.linear_model  import LogisticRegression
from sklearn.metrics import accuracy_score
lr=LogisticRegression()
lr.fit(X_train_nor,y_train)

y_pred = lr.predict(X_test_nor)
print ("So, Our accuracy Score is: {}".format(round(accuracy_score(y_pred, y_test),4)))
```
