```python
import pandas as pd
import re
```

```python
data = pd.read_csv('test3.csv')
```

```python
data
```

```python
secondDuration = []
```

```python
#transfer chinese time to seconds
for time in data['duration']:
    timeList = time.replace('分',' ').replace('秒',' ').split()
    secondsTime = int(timeList[0]) * 60 + int(timeList[1])
    secondDuration.append(secondsTime)
```

```python
data['duration'] = secondDuration
data
```

```python
#transfer chinese result 勝利->1 敗-> 0
data['result'].replace()
data = data.replace(to_replace = ['勝利','敗'],value = ['1','0'])
data
```

```python
#Inbalanced distribution status
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,5))
data.result.value_counts(normalize = True).plot(kind='bar', color= ['skyblue','navy'], alpha = 0.9, rot=0)
plt.title('Game result Lose(0) and Win(1) in the Imbalanced Dataset')
plt.show()
```

```python
#count 1 and 0 frequency in result
count0 = 0
count1 = 0
for i in range(len(data)):
    if data.loc[i]['result'] == '0':
        count0 += 1
    else:
        count1 += 1
print(count0)
print(count1)
```

```python
#Balance data
for i in range(len(data)):
    if count1 > 2503:
        break
    if data.loc[i]['result'] == '0':
        temp = data.loc[i]['team_1']
        data.loc[i, 'team_1'] = data.loc[i]['team_2']
        data.loc[i, 'result'] = '1'
        data.loc[i, 'team_2'] = temp
        count1 += 1
        count0 -= 1
```

```python
data
```

```python
#Balanced distribution status
fig = plt.figure(figsize = (8,5))
data.result.value_counts(normalize = True).plot(kind='bar', color= ['skyblue','navy'], alpha = 0.9, rot=0)
plt.title('Game result Lose(0) and Win(1) in the Imbalanced Dataset')
plt.show()
```

```python
top1List = []
jg1List = []
mid1List = []
ad1List = []
sup1List = []
top2List = []
jg2List = []
mid2List = []
ad2List = []
sup2List = []
```

```python
#split team1 champions
for team1 in data['team_1']:
    team1List = team1.replace(',',' ').split()
    top1,jg1,mid1,ad1,sup1 = team1List
    top1List.append(top1)
    jg1List.append(jg1)
    mid1List.append(mid1)
    ad1List.append(ad1)
    sup1List.append(sup1)
```

```python
data['top1'] = top1List
data['jg1'] = jg1List
data['mid1'] = mid1List
data['ad1'] = ad1List
data['sup1'] = sup1List
del data['team_1']
data
```

```python
#split team2 champions
for team2 in data['team_2']:
    team2List = team2.replace(',',' ').split()
    top2,jg2,mid2,ad2,sup2 = team2List
    top2List.append(top2)
    jg2List.append(jg2)
    mid2List.append(mid2)
    ad2List.append(ad2)
    sup2List.append(sup2)
```

```python
data['top2'] = top2List
data['jg2'] = jg2List
data['mid2'] = mid2List
data['ad2'] = ad2List
data['sup2'] = sup2List
del data['team_2']
data
```

```python
#Classify
top1type = []
jg1type = []
mid1type = []
ad1type = []
sup1type = []
top2type = []
jg2type = []
mid2type = []
ad2type = []
sup2type = []
for i in range(len(data)):
    for j in charactersDict:
        types = charactersDict[j]
        if data.loc[i]['top1'] in types:
            top1type.append(j)
        if data.loc[i]['jg1'] in types:
            jg1type.append(j)
        if data.loc[i]['mid1'] in types:
            mid1type.append(j)
        if data.loc[i]['ad1'] in types:
            ad1type.append(j)
        if data.loc[i]['sup1'] in types:
            sup1type.append(j)
        if data.loc[i]['top2'] in types:
            top2type.append(j)
        if data.loc[i]['jg2'] in types:
            jg2type.append(j)
        if data.loc[i]['mid2'] in types:
            mid2type.append(j)
        if data.loc[i]['ad2'] in types:
            ad2type.append(j)
        if data.loc[i]['sup2'] in types:
            sup2type.append(j)
```

```python
data['top1type'] = top1type
data['jg1type'] = jg1type
data['mid1type'] = mid1type
data['ad1type'] = ad1type
data['sup1type'] = sup1type
data['top2type'] = top2type
data['jg2type'] = jg2type
data['mid2type'] = mid2type
data['ad2type'] = ad2type
data['sup2type'] = sup2type
data
```
