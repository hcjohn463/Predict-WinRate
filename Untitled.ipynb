```python
import pandas as pd
import re
```

```python
data = pd.read_csv('test3.csv')
```

```python
data
```

```python
secondDuration = []
```

```python
#transfer chinese time to seconds
for time in data['duration']:
    timeList = time.replace('分',' ').replace('秒',' ').split()
    secondsTime = int(timeList[0]) * 60 + int(timeList[1])
    secondDuration.append(secondsTime)
```

```python
data['duration'] = secondDuration
data
```

```python
#transfer chinese result 勝利->1 敗-> 0
data['result'].replace()
data = data.replace(to_replace = ['勝利','敗'],value = ['1','0'])
data
```

```python
top1List = []
jg1List = []
mid1List = []
ad1List = []
sup1List = []
top2List = []
jg2List = []
mid2List = []
ad2List = []
sup2List = []
```

```python
#split team1 champions
for team1 in data['team_1']:
    team1List = team1.replace(',',' ').split()
    top1,jg1,mid1,ad1,sup1 = team1List
    top1List.append(top1)
    jg1List.append(jg1)
    mid1List.append(mid1)
    ad1List.append(ad1)
    sup1List.append(sup1)
```

```python
data['top1'] = top1List
data['jg1'] = jg1List
data['mid1'] = mid1List
data['ad1'] = ad1List
data['sup1'] = sup1List
del data['team_1']
data
```

```python
#split team2 champions
for team2 in data['team_2']:
    team2List = team2.replace(',',' ').split()
    top2,jg2,mid2,ad2,sup2 = team2List
    top2List.append(top2)
    jg2List.append(jg2)
    mid2List.append(mid2)
    ad2List.append(ad2)
    sup2List.append(sup2)
```

```python
data['top2'] = top2List
data['jg2'] = jg2List
data['mid2'] = mid2List
data['ad2'] = ad2List
data['sup2'] = sup2List
del data['team_2']
data
```
